```python
alimentadores = []
for i in df['Identificador']:
    if i not in alimentadores:
        alimentadores.append(i)
```

```python
result = pd.DataFrame()
for i in range(len(alimentadores)):
    df_sum = df.loc[df['Identificador'] == alimentadores[i]]
    df_sum = df_sum.reset_index(drop=True)
    result[alimentadores[i]] = df_sum['Potencia Ativa Consistida (kW)']
    
    
    
```

```python
result['total'] = result.apply(lambda row: row.sum(), axis=1)
```
